```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
def plots(file):
    f = open(file)
    data = json.load(f)
    f.close()

    bestFitness = np.array(data['bestFitness'])
    averageFitness = np.array(data['averageFitness'])

    bestVisibility = np.array(data['bestVisibility'])

    bestLocations = np.array(data['bestLocations'])
    averageLocations = np.array(data['averageLocations'])

    standardDeviation = np.array(data['standardDeviation'])
    #standardDeviation[0] = standardDeviation[1]

    x = np.arange(len(bestFitness))

    plt.plot(x, bestFitness, c='black', label='fitness')
    #plt.axhline(y = bestFitness.max(), c='black', linestyle='--')
    #plt.text(len(bestFitness) * 1.1, bestFitness.max(), s=str(bestFitness.max()))
    plt.text(len(bestFitness) * 1.1, bestFitness[-1], s=str(bestFitness[-1]))
    plt.xlabel('generations')
    plt.ylabel('fitness')
    plt.title('Fitness')
    plt.show()

    plt.plot(x, bestVisibility, c='black', label='visibility')
    #plt.axhline(y = bestVisibility.max(), c='black', linestyle='--')
    #plt.text(len(bestVisibility) * 1.1, bestVisibility.max(), s=str(bestVisibility.max()))
    #plt.axhline(y = bestVisibility[-1], c='black', linestyle='--')
    plt.text(len(bestVisibility) * 1.1, bestVisibility[-1], s=str(bestVisibility[-1]))
    plt.xlabel('generations')
    plt.ylabel('visibility')
    plt.title('Visibility')
    plt.show()

    plt.plot(x, bestLocations, c='black', label='best')
    #plt.plot(x, averageLocations, c='black', linestyle='--', label='average')
    plt.axhline(y = bestLocations.max(), c='black', linestyle='--')
    plt.text(len(bestLocations) * 1.1, bestLocations.max(), s=str(bestLocations.max()))
    plt.axhline(y = bestLocations[-1], c='black', linestyle='--')
    plt.text(len(bestLocations) * 1.1, bestLocations[-1], s=str(bestLocations[-1]))
    plt.xlabel('generations')
    plt.ylabel('locations')
    plt.title('Locations')
    plt.show()

    #plt.plot(x, standardDeviation, c='black', label='σ')
    #plt.axhline(y = bestVisibility.max(), c='black', linestyle='--')
    #plt.text(len(bestVisibility) * 1.1, bestVisibility.max(), s=str(bestVisibility.max()))
    #plt.xlabel('generations')
    #plt.ylabel('σ')
    #plt.title('Standard deviation')
    #plt.show()


    ## 95% rule
    #p95 = np.array([0.0] * len(x))
    #for i in range(len(x)):
    #    p95[i] = averageFitness[i] + 2 * standardDeviation[i]

    #plt.plot(x, p95, label='95%')
    #plt.plot(x, bestFitness, label='best')
    #plt.plot(x, averageFitness, label='average')
    #plt.legend()
    #plt.show()
```
