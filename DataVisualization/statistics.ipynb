```python
import matplotlib.pyplot as plt
import numpy as np
import json
from scipy import stats
```

```python
def fourhist(a, b, c, d, xlabel, ylabel):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)

    n, x, _ = ax1.hist(a, density=True)
    density = stats.gaussian_kde(a)
    ax1.plot(x, density(x))
    ax1.set_ylabel(ylabel)

    if stats.shapiro(a)[1] > 0.05:
        print('normal')
        print(np.std(a))
    else:
        print('not normal')

    n, x, _ = ax2.hist(b, density=True)
    density = stats.gaussian_kde(b)
    ax2.plot(x, density(x))

    if stats.shapiro(b)[1] > 0.05:
        print('normal')
        print(np.std(b))
    else:
        print('not normal')

    if c is not None:
        n, x, _ = ax3.hist(c, density=True)
        density = stats.gaussian_kde(c)
        ax3.plot(x, density(x))
        ax3.set_xlabel(xlabel)
        ax3.set_ylabel(ylabel)

        if stats.shapiro(c)[1] > 0.05:
            print('normal')
            print(np.std(c))
        else:
            print('not normal')
    else:
        ax3.axis('off')
        ax1.set_xlabel(xlabel)

    if d is not None:
        n, x, _ = ax4.hist(d, density=True)
        density = stats.gaussian_kde(d)
        ax4.plot(x, density(x))
        ax4.set_xlabel(xlabel)

        if stats.shapiro(d)[1] > 0.05:
            print('normal')
            print(np.std(d))
        else:
            print('not normal')
    else:
        ax4.axis('off')
        ax2.set_xlabel(xlabel)

    plt.show()
```
