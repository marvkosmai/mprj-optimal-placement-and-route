```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

# Sampling Pseudo Random vs Van der Corput

```python
f = open('data/sampling.json')
data = json.load(f)
f.close()

labels = data['samples']
pseudo_random = data['pseudoRandom']
van_der_corput = data['vanDerCorput']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, pseudo_random, width, label='Pseudo Random')
rects2 = ax.bar(x + width/2, van_der_corput, width, label='Van der Corput')

# Add some text for labels, title and custom x-axis tick labels, etc.

ax.set_ylabel('Sum of all absolut differences (1)')
ax.set_xlabel('Number of samples')
ax.set_title('Difference by number of samples and method')
ax.set_xticks(x, labels)
ax.legend()

#ax.bar_label(rects1, padding=3)
#ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
f = open('data/sampling.json')
data = json.load(f)
f.close()

labels = data['samplesStr']
pseudo_random = data['pseudoRandom']
van_der_corput = data['vanDerCorput']

fig = plt.figure()

ax = fig.add_subplot()
ax.plot(labels, pseudo_random, 'ko-', label="Pseudorandom")
ax.plot(labels, van_der_corput, 'ko--' , label="Van der Corput")

ax.set_ylabel('Sum of all absolut differences (1)')
ax.set_xlabel('Number of samples')
ax.set_title('Difference by number of samples and method')
ax.xaxis.set_ticks(labels)
ax.xaxis.set_ticklabels(labels)
ax.legend()

plt.show()
```
