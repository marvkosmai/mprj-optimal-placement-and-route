```python
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
def bestLocAndBestVis(path):
    f = open(path)
    data = json.load(f)
    f.close()

    bl = np.array(data['bestLocations'])
    bv = np.array(data['bestVisibility'])
    av = np.array(data['averageVisibility'])
    al = np.array(data['averageLocations'])

    steps = 1000
    bl = bl[:steps]
    bv = bv[:steps]
    av = av[:steps]
    al = al[:steps]

    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)

    x = np.arange(len(bl))

    ax1.plot(x, bl, color='black')

    ax2.plot(x, bv, color='black')

    ax3.plot(x, al, color='black')

    ax4.plot(x, av, color='black')

    plt.show()
```

```python
def allFromOne(path1, path2, path3, path4, category, label):
    f = open(path1)
    data1 = json.load(f)
    f.close()

    f = open(path2)
    data2 = json.load(f)
    f.close()

    f = open(path3)
    data3 = json.load(f)
    f.close()

    f = open(path4)
    data4 = json.load(f)
    f.close()

    steps = 1000

    bl1 = np.array(data1[category])[:steps]
    bl2 = np.array(data2[category])[:steps]
    bl3 = np.array(data3[category])[:steps]
    bl4 = np.array(data4[category])[:steps]

    x = np.arange(len(bl1))

    plt.plot(x, bl1, linestyle='dotted', color='black', label='2X1')
    plt.plot(x, bl2, linestyle='dashdot', color='black', label='X2')
    plt.plot(x, bl3, linestyle='dashed', color='black', label='C6')
    plt.plot(x, bl4, linestyle='solid', color='black', label='D7')

    plt.xlabel('Generations')
    plt.ylabel(label)

    plt.legend()

    plt.show()
```

```python
def allFromOneModel(path1, path2, path3, path4, category, label):
    f = open(path1)
    data1 = json.load(f)
    f.close()

    f = open(path2)
    data2 = json.load(f)
    f.close()

    f = open(path3)
    data3 = json.load(f)
    f.close()

    f = open(path4)
    data4 = json.load(f)
    f.close()

    steps = 1000

    bl1 = np.array(data1[category])[:steps]
    bl2 = np.array(data2[category])[:steps]
    bl3 = np.array(data3[category])[:steps]
    bl4 = np.array(data4[category])[:steps]

    x = np.arange(len(bl1))

    plt.plot(x, bl1, linestyle='dotted', color='black', label='Wall')
    plt.plot(x, bl2, linestyle='dashdot', color='black', label='Tree')
    plt.plot(x, bl3, linestyle='dashed', color='black', label='Capsule')
    plt.plot(x, bl4, linestyle='solid', color='black', label='Cube')

    plt.xlabel('Generations')
    plt.ylabel(label)

    plt.legend()

    plt.show()
```
